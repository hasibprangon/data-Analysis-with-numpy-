```python
import numpy as np

arr_1 = np.array([1,2,3])

arr_1 = arr_1 * 5
print(arr_1)
```

## Ndarray

```python
import numpy as np

# 1D array
arr1 = np.array([1,2,3,4,5])

print(type(arr1))
#show dimention
print(arr1.ndim)

# 2d array

arr2 = np.array([ [1,2,3], #row-1
                  [4,5,6] #row-2
                ])

print(arr2.ndim)


#3d array  

arr3 = np.array([
#first floor
[ [1,2,3], #row-1
[4,5,6] ] ,#row-2
                
#2nd floor

[ [7,8,9], #row-1
[10, 11, 12.5] ] #row-2

])

print(arr3.ndim)
```

### array atributes

```python
#dimention
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)


#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)


#data types
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)


#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```
