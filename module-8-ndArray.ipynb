## Ndarray

```python
import numpy as np

# 1D array
arr1 = np.array([1,2,3,4,5])

print(type(arr1))
#show dimention
print(arr1.ndim)

# 2d array

arr2 = np.array([ [1,2,3], #row-1
                  [4,5,6] #row-2
                ])

print(arr2.ndim)


#3d array  

arr3 = np.array([
#first floor
[ [1,2,3], #row-1
[4,5,6] ] ,#row-2
                
#2nd floor

[ [7,8,9], #row-1
[10, 11, 12.5] ] #row-2

])

print(arr3.ndim)
```

### array atributes

```python
#dimention
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)


#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)


#data types
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)


#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

### ndarray data type

```python
arr = np.array([1,2,3])

print(arr.dtype)


#upcasting
#upcasted to float
arr = np.array([1,2,3.5])
#upcasted to string
arr = np.array([1,2,3.5, '3'])
print(arr.dtype)
```

### selecting a data type for an array

```python
arr = np.array([1,2,3], dtype=np.int8)
# arr = np.array([1,2,300], dtype=np.int8)
arr = np.array([1,2,300], dtype=np.int16)
print(arr.dtype)
print(arr)
```

```python
arr = np.array([1,2,300])
print(arr.dtype)
arr = arr.astype(np.int16)
print(arr.dtype)
```

```python
arr = np.array([1,2,300], dtype=np.int16)
print(arr.dtype)
print(arr)

arr = arr.astype(np.int32)
print(arr.dtype)


#error
# arr = np.array([1,2,3.5, 'hello'], dtype= np.float64)
```

## ndarray creation from existing data

```python
#tuple

tpl = (1,2,3)

arr = np.array(tpl)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#set

st = {1,2,3}

arr = np.array(list(st))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#dictionary 

dc = {
    'a' : 10,
    'b' : 20,
    'c' : 30
}

keys = dc.keys()
values = dc.values()
items = dc.items()

#keys to ndarray
# arr = arr.astype(np.int16)
arr = np.array(list(keys))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)

# print(keys)
# print(values)
# print(items)
```

```python
# values to ndarray

arr = np.array(list(values))
arr = arr.astype(np.int16)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

```

```python
#items to ndarray

arr = np.array(list(items))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## array creation with range function

```python
#np.arange(first, end, step) -> similar to range function

arr = np.arange(1, 10, 1).reshape(3,3)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#np.linspace(start, stop, num)

arr = np.linspace(1, 10, 15)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#np.logspace(start, stop, num, base=10)

arr = np.logspace(0, 4, 5)
print(arr)
```
